# Prepare

```python
# drop non-quantitative and irrelivent data
df = df[['race', 'q08_hispanic', 'income', 'pregnancy_stage', 'q01_overall',
       'q02_financial','q06_birthdoula','q06_postpartumdoula', 
       'q06_lactation', 'q06_mentalhealth', 'q06_social', 
       'q06_diapers', 'q06_formula', 'q06_food', 
       'q06_housing','q06_transportation', 'q06_healthinsurance', 
       'q06_financial','q06_covid19education', 'q06_support_none', 
       'q06_support_other']]
df.head()
```

```python
# change data types of overall and financial to objects
df = df.astype(object)
df.info()
```

```python
# change responces of overall and to catagories

df.q01_overall = df.q01_overall.apply(lambda number : str(number).replace('1.0','We\'ve got this!'))
df.q01_overall = df.q01_overall.apply(lambda number : str(number).replace('2.0','We\'re doing OK.'))
df.q01_overall = df.q01_overall.apply(lambda number : str(number).replace('3.0','So-So.'))
df.q01_overall = df.q01_overall.apply(lambda number : str(number).replace('4.0','It\'s a struggle.'))
df.q01_overall = df.q01_overall.apply(lambda number : str(number).replace('5.0','Barely making it.'))

df.q02_financial = df.q02_financial.apply(lambda number : str(number).replace('1.0','We\'ve got this!'))
df.q02_financial = df.q02_financial.apply(lambda number : str(number).replace('2.0','We\'re doing OK.'))
df.q02_financial = df.q02_financial.apply(lambda number : str(number).replace('3.0','So-So.'))
df.q02_financial = df.q02_financial.apply(lambda number : str(number).replace('4.0','It\'s a struggle.'))
df.q02_financial = df.q02_financial.apply(lambda number : str(number).replace('5.0','Barely making it.'))
                                                                          
print(df.q01_overall.value_counts())
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print(df.q02_financial.value_counts())
```

```python
# fixing value mismatch 

df.income = df.income.apply(lambda number : str(number).replace('$40,000 or under','40,000 and under'))

df.income.value_counts()
```

```python
# Replace nans with "Zero Response"
df = df.fillna(value="Zero Response")
df.head()
```

```python
df.income.value_counts()
```

```python
# create non-white column for respondance who identify as non-white
df.nonwhite = (df.race != "White") & (df.race != "Undisclosed")

df.nonwhite.value_counts()
```
